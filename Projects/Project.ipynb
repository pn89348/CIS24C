```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
import nltk
nltk.download('vader_lexicon')
```

```python
def sentiment_score(review):
    '''This function inputs a review and returns the sentiment score (between -1 to 1)'''
    review = str(review)
    sentiment_analyzer = SentimentIntensityAnalyzer()
    sentiment_dict = sentiment_analyzer.polarity_scores(review)
    return sentiment_dict['compound']
```

```python
# Testing the function
sentiment_score("I am frustrated with kind of service I received on this case")
```

```python
# Creating a new column with the sentiment score
df["Sentiment Score"] = df["reviews.text"].apply(sentiment_score)
```

```python
# Getting the min and max sentiment
min_sentiment = df["Sentiment Score"].min()
max_sentiment = df["Sentiment Score"].max()
```

```python
def normalize(sentiment):
    '''This function inputs a sentiment score and normalizes it to be between 1 and 5'''
    result = 4 * ((sentiment - min_sentiment) / (max_sentiment - min_sentiment)) + 1
    return result
```

```python
# Testing the function
normalize(0)
```

```python
# Creating a new column with the normalized sentiment scores
df["Normalized Sentiment Score"] = df["Sentiment Score"].apply(normalize)
```

```python
def round_normalize(sentiment):
    '''This function inputs a sentiment score and normalizes and rounds it to be an integer between 1 and 5'''
    result = round(normalize(sentiment))
    return result
```

### Regression

```python
# Importing NumPy library (for advanced math operations)
import numpy as np
```

```python
def regression(x, y):
    '''This is a function that inputs x and y and calculates the variables m and b'''
    # xbar and ybar
    xbar = np.mean(x)
    ybar = np.mean(y)
    
    # Calculating Regression
    m = np.sum((x - xbar) * (y - ybar)) / np.sum(np.square(x - xbar))
    b = ybar - m * xbar
    
    return m, b
```

```python
x = np.array(df['Normalized Sentiment Score'])
y = np.array(df['reviews.rating'])
```

```python
print(regression(x, y))
```

```python
# Getting the values of m and b from the function
m = regression(x, y)[0]
b = regression(x, y)[1]
```

```python
def mxb(x):
    '''This is a function that calculates mx+b'''
    result = m * x + b
    return result
```

```python
# Creates a new column with the mx+b coeffecient in it
df["Regression Rating"] = df["Normalized Sentiment Score"].apply(mxb)
```

### Correlation

```python
def correlation(x, y):
    '''This function inputs x and y and returns the correlation'''
    # xbar and ybar
    xbar = np.mean(x)
    ybar = np.mean(y)
    
    # Calculating Correlation with Pearson Corrrelation Coefficient
    correlation = np.sum((x - xbar) * (y - ybar)) / (np.sqrt(np.sum(np.square(x - xbar))) * np.sqrt(np.sum(np.square(y - ybar))))
    
    return correlation
```

```python
# Testing the Function
print(correlation(np.array([1, 2, 3, 4]), np.array([2, 4, 6, 8])))
print(correlation(np.array([1, 20920289489, -68904, 111111]), np.array([43233333, 664543, 45345, 5555])))
print(correlation(np.array([1, 2, 3, 4]), np.array([-2, -4, -6, -8])))
```

### Happy or Unhappy

```python
def is_happy_pred(review):
    '''This function inputs a review and checks if the user is happy, sad, or nuetral depending on the sentiment it calculates'''
    review = str(review)
    sentiment_analyzer = SentimentIntensityAnalyzer()
    sentiment_dict = sentiment_analyzer.polarity_scores(review)
    score = sentiment_dict['compound']
    normalized_score = normalize(score)
    if normalized_score >= 3.5:
        return True
    elif normalized_score < 2.5:
        return False
    else:
        return None
```

```python
def is_happy(rating):
    '''This function inputs the rating the user gave and checks if they are happy based on that'''
    if rating >= 3.5:
        return True
    elif rating < 2.5:
        return False
    else:
        return None
```

```python
# Creates two new columns with the prediction of whether the user is happy or not (based on sentiment) and the actual happiness based on the rating the user gave 
df["Is happy Prediction"] = df["reviews.text"].apply(is_happy_pred)
df["Is happy Actual"] = df["reviews.rating"].apply(is_happy)
```
